```python
import matplotlib.pyplot as plt
from scipy.linalg import solve
```

```python
def f(x):
    return(np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2))
```

## ÐŸÐ¾Ð»Ð¸Ð½Ð¾Ð¼ Ð²Ñ‚Ð¾Ñ€Ð¾Ð¹ ÑÑ‚ÐµÐ¿ÐµÐ½Ð¸

```python
X=np.array([[1,1,1], [1,8,8**2], [1,15,15**2]])
y=np.array([f(1), f(8), f(15)])
```

```python
w=solve(X,y)
```

```python
def poly(w,x):
    x=np.array([1,x,x**2])
    return(np.sum(w*x))
```

```python
plt.plot(np.arange(1,16, 0.5), f(np.arange(1,16, 0.5)))
plt.plot(np.arange(1,16, 0.5), poly(w,np.arange(1,16, 0.5)))
```
